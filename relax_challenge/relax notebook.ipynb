```python
engagement['time_stamp'] = pd.to_datetime(engagement['time_stamp'])
```

```python
engagement_shifted = engagement.shift(2)
```

```python
engagement_shifted.columns = ['time_stamp_shifted','user_id_shifted','visited_shifted']
```

```python
engagement = pd.concat([engagement, engagement_shifted], axis=1)
```

```python
engagement
```

```python
engagement.info()
```

```python
engagement = engagement.iloc[2:,]
engagement = engagement.astype({'user_id_shifted':'int','visited_shifted':'int'})
```

```python
engagement
```

```python
engagement.info()
```

```python
engagement['time_dif'] = engagement['time_stamp']-engagement['time_stamp_shifted']
```

```python
engagement = engagement[engagement['user_id']==engagement['user_id_shifted']]
```

```python
engagement = engagement[engagement['time_dif']<='7 days']
```

```python
active_users = set(engagement['user_id'])
```

```python
users['adopted_user'] = [ int(x in active_users ) for x in users['object_id']]
```

```python
features_at_sign_up=users[['creation_source','opted_in_to_mailing_list','enabled_for_marketing_drip','org_id','invited_by_user_id','adopted_user']]
```

```python
features_at_sign_up
```

```python
features_at_sign_up.org_id.value_counts()
```

with 417 seperate organizations that would be a lot of fields to test over

```python
features_at_sign_up['invited_by_adopted_user'] = [0 if pd.isna(x) else features_at_sign_up.iloc[int(x)-1,5] for x in features_at_sign_up['invited_by_user_id']]
```

```python
features_at_sign_up = features_at_sign_up.drop(columns=['invited_by_user_id'])
```

```python
features_at_sign_up
```

```python
features_at_sign_up.groupby(['org_id']).mean().sort_values(by=['adopted_user'])
```

What orgaization a user is from strongly determines probaility of adopted user status. Some orgs have no adopted users while others have over 30%.

```python
features_at_sign_up.groupby(['invited_by_adopted_user']).mean().sort_values(by=['adopted_user'])
```

A user has a 13% chance of being an adopted user if they are not invited by an adopted user. They have a 21% of being an adopted user if they are invited by an adopted user

```python
features_at_sign_up.groupby(['creation_source']).mean().sort_values(by=['adopted_user'])
```

How one signed up has an influence on use. Users invited by a guest to an organization or signed up with a google authorization have a 17% chance of being adopted users. People who signedup to another person's personal project have an 8% chance.

```python
features_at_sign_up.groupby(['opted_in_to_mailing_list']).mean().sort_values(by=['adopted_user'])
```

There is only a slight difference between those who did and did not sign-up for the mailing list.

```python
features_at_sign_up.groupby(['enabled_for_marketing_drip']).mean().sort_values(by=['adopted_user'])
```

likewise, marketing-drip-enabled users not substantially more likely to become adopted users.

Conclusions: The best indicators for if a user will become an adopted user from initial log-in information are 1)the org that they are a member of and 2) if they were invited by an adopted user.
